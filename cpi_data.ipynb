```python
import pandas as pd
```

```python
#Read xlsx data file
cpi = pd.read_excel("Resources/CPI.xlsx", sheet_name="Data1")

cpi_df=pd.DataFrame(cpi)

#Remove unwanted rows
cpi_data_df = cpi_df.drop(labels=range(0, 147), axis=0)
cpi_data_df.head()
```

```python
cpi_data_df.columns
```

```python
# Remove columns
cpi_updated_df = cpi_data_df[["Unnamed: 0", "Index Numbers ;  All groups CPI ;  Australia ;", 
                   "Percentage Change from Previous Period ;  All groups CPI ;  Australia ;"]]

cpi_updated_df.head()
```

```python
# Rename columns
clean_cpi_data_df = cpi_updated_df.rename(columns={"Unnamed: 0":"Date",
                                                "Index Numbers ;  All groups CPI ;  Australia ;": "CPI Australia",
                                                "Percentage Change from Previous Period ;  All groups CPI ;  Australia ;":"% Change CPI Australia"})

clean_cpi_data_df.head()
```

```python
clean_cpi_data_df["Date"] = pd.to_datetime(clean_cpi_data_df.Date)
clean_cpi_data_df.head()
```

```python
clean_cpi_data_df.set_index("Date", inplace=True)
clean_cpi_data_df
```
